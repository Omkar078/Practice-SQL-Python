```python
## TO find second highest runner up score from the list of scores

n = [2, 3, 6, 6, 5]
scores = n 
unique_scores = list(set(scores))
unique_scores.sort()
print(unique_scores[-2])
```

```python
## To check  negative number is present in the  list
numbers = [1, 2, 3, -4, 5]
has_negative = any(num < 0 for num in numbers)
print(has_negative)  # Output: True
```

```python
### how many negative numbers are there in the list
n = 0
for n in numbers:
    if n < 0:
        print("Negative number found", " : ", n)
        break
```

```python
### how many negative numbers are there in the list
count = 0
for n in numbers:
    if n < 0:
        count += 1
print("Total negative numbers:", count)
```

```python
### check if all values in the list are intergers
values = [1, 2, 3, 4, 5]
all_integers = all(isinstance(value, int) for value in values)
print(all_integers)  # Output: True
```

```python
### All elements in the list are same or not
lst = [1, 1, 1, 1]
print(len(set(lst)) == 1)  # Output: True
```

```python
###code to find repetitive characters  in a list
char_list = ['a', 'b', 'c', 'a', 'd', 'b', 'e']
repetitive_chars = set()
for char in char_list:
    if char_list.count(char) > 1:
        repetitive_chars.add(char)
print(repetitive_chars)  # Output: {'a', 'b'}
```

```python
###code to find repetitive numbers  in a list
n =  [4, 2, 4, 3, 2, 1, 3, 5]
repetitive_numbers = set()
for num in n:
    if n.count(num) > 1:
        repetitive_numbers.add(num)
print(repetitive_numbers)  # Output: {2, 3, 4}
```

```python
###code to find unique numbers in a list preserving the order
uniq_num = []
for n in n:
    if n not in uniq_num:
        uniq_num.append(n)
        
print(uniq_num)
```

```python
#To find sum of elements in a list
numbers = [1, 2, 3, 4, 5]
n = []
for num in numbers:
    n.append(num)
print(sum(n))  # Output: 15
```

```python
### To find sum of even numbers in a list
numbers = [1, 2, 3, 4, 5, 6]
even_sum = 0    
for num in numbers:
    if num % 2 == 0:
        even_sum += num
print("Sum of even numbers:", even_sum)  # Output: 12
```

```python
### To find sum of  odd numbers in a list
numbers = [1, 2, 3, 4, 5, 6]
odd_sum = 0     
for num in numbers:
    if num % 2 != 0:
        odd_sum += num
print("Sum of odd numbers:", odd_sum)  # Output: 9
```
